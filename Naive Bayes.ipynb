```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv("placement._Data.csv")
```

```python
dataset.head(3)
```

```python
sns.kdeplot(data=dataset["cgpa"])
```

```python
sns.kdeplot(data=dataset["iq"])
```

```python
dataset.isnull().sum()
```

```python
sns.scatterplot(x="cgpa",y="iq",data=dataset,hue="placement")
```

```python
x=dataset.iloc[:,:-1]
y=dataset["placement"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

GaussianNB

```python
gnb=GaussianNB()
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test)*100 ,gnb.score(x_train,y_train)*100 
```

```python
gnb.predict([[5.9	,106	]])
```

```python
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gnb)
plt.show()
```

MultinomialNB

```python
mnb=MultinomialNB()
mnb.fit(x_train,y_train)
```

```python
mnb.score(x_test,y_test)*100 ,mnb.score(x_train,y_train)*100 
```

```python
mnb.predict([[5.9	,106	]]) #wrong prediction 
```

```python
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=mnb)
plt.show()
```

BernoulliNB

```python
bnb=BernoulliNB()
bnb.fit(x_train,y_train)
```

```python
bnb.score(x_test,y_test)*100 ,bnb.score(x_train,y_train)*100 
```

```python
bnb.predict([[5.9	,106	]])
```

```python
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=bnb)
plt.show
```

GaussianNB is the right model for this dataset because it best fit for continous data
when the feature are in normal distribution
