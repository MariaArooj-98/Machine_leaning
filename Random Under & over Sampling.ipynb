```python
import pandas as pd
```

```python
dataset=pd.read_csv("Social_Network_Ads.csv")
dataset.head(10)
```

```python
x=dataset.iloc[:,:-1]
y=dataset["Purchased"]
```

```python
dataset["Purchased"].value_counts()
```

```python
from imblearn.under_sampling import RandomUnderSampler
```

Random Under Sampling 

```python
Ru =RandomUnderSampler()
ru_x,ru_y=Ru.fit_resample(x,y)
```

```python
ru_y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(ru_x, ru_y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[32,150000]])
```

Random Over Sampling

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ro =RandomOverSampler()
ro_x,ro_y=Ru.fit_resample(x,y)
```

```python
ro_y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train1, x_test1, y_train1, y_test1 = train_test_split(ro_x, ro_y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
lr1=LogisticRegression()
lr1.fit(x_train1,y_train1)
```

```python
lr1.score(x_test1,y_test1)*100
```

```python
lr1.predict([[46,28000]])
```
