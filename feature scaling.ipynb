```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("loan_data_set.csv")
dataset.head(3)
```

```python
sns.displot(dataset["ApplicantIncome"])
plt.show()
```

```python
#feature Scaling
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss= StandardScaler()
ss.fit(dataset[["ApplicantIncome"]])
```

```python
dataset["ApplicantIncome_ss"]=pd.DataFrame(ss.transform(dataset[["ApplicantIncome"]]), columns=["x"])
```

```python
dataset.head(3)
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(10,7))

plt.title("Before")
sns.displot(dataset["ApplicantIncome_ss"])
```

```python

plt.title("After")
sns.displot(dataset["ApplicantIncome"])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mms= MinMaxScaler()
mms.fit(dataset[["CoapplicantIncome"]])
```

```python
dataset["CoapplicantIncome_minmax"]=pd.DataFrame(mms.transform(dataset[["CoapplicantIncome"]]), columns=["x"])
```
